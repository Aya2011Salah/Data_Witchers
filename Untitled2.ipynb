```python
import pandas as pd
from prophet import Prophet

df = pd.read_csv('/sales store cleaned.csv', encoding='latin-1')

model = Prophet(daily_seasonality=True)

print(df.columns)

df['Order Date'] = pd.to_datetime(df['Order Date'])
grouped_df = df.groupby('Order Date').sum()


total_order_per_day = df.groupby('Order Date')['Order ID'].count().reset_index()
total_order_per_day = total_order_per_day.rename(columns={'Order Date': 'ds', 'Order ID': 'y'})

model.fit(total_order_per_day)
future = model.make_future_dataframe(periods=30)
forecast = model.predict(future)
forecast_df = forecast[['ds', 'yhat']]
forecast_df.tail(30)

model.plot(forecast)
```
